```python
from dotenv import load_dotenv  
load_dotenv()

#Create an Anthropic API
from anthropic import Anthropic
client = Anthropic()
model = "claude-sonnet-4-6"

#make a request
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages,system=None,temperature=1.0, stop_sequences=None):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature
        }
    if system:
        params["system"] = system
    if stop_sequences:
        params["stop_sequences"] = stop_sequences

    message = client.messages.create(**params)
    return message.content[0].text
```

```python
messages = []

add_user_message(messages, "Generate a very short event bridge rule as json. Output only the JSON, nothing else.")
add_assistant_message(messages, "{")  # Prefill: forces model to continue from "{" — no markdown fences

text = "{" + chat(messages)  # Prepend the prefilled "{" since it's not included in the response
print(text)
```

```python
import json

print(repr(text))  # Debug: show exact contents of text
parsed = json.loads(text.strip())
print(parsed)
```

```python
messages = []

add_user_message(messages, "Give me three useful bash commands.")
add_assistant_message(messages, "1. ")  # Prefill starts the numbered list — model must continue from here

text = chat(messages, stop_sequences=["\n4."])  # Stop before a 4th item could appear

print("1. " + text)
```
