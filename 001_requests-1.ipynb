```python
# Load environment variables
from dotenv import load_dotenv  
load_dotenv()
```

```python
#Create an Anthropic API
from anthropic import Anthropic
client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
# Test API key
try:
    response = client.messages.create(
        model="claude-sonnet-4-6",  # Try a known model
        max_tokens=10,
        messages=[{"role": "user", "content": "Hello"}]
    )
    print("вњ… API key is working!")
    print("Response:", response.content[0].text)
except Exception as e:
    if "authentication" in str(e).lower() or "401" in str(e):
        print("вќЊ API key is invalid:", str(e))
    elif "not_found" in str(e).lower() or "404" in str(e):
        print("вњ… API key is working, but model not found. Try a different model.")
        print("Error:", str(e))
    else:
        print("вќ“ Unexpected error:", str(e))
```
